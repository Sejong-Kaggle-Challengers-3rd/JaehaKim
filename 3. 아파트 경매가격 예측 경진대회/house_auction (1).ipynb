```python
import pandas as pd
import numpy as np
train=pd.read_csv('/content/Auction_master_train.csv')
test=pd.read_csv('/content/Auction_master_test.csv')
regist=pd.read_csv('/content/Auction_regist.csv')
rent=pd.read_csv('/content/Auction_rent.csv')
result=pd.read_csv('/content/Auction_result.csv')
train=train.drop(['addr_li','Specific'],axis=1)
train
```

```python
col=[]
for column in train.columns:
  if train[column].dtypes=='O':
    col.append(column)
print(col)
#형식이 object인 column들을 알아냄

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for c in col:
  train[c]=le.fit_transform(train[c])

print(train.dtypes)
```

```python
train['Appraisal_date']=pd.to_datetime(train['Appraisal_date'])
train['First_auction_date']=pd.to_datetime(train['First_auction_date'])
train['Final_auction_date']=pd.to_datetime(train['Final_auction_date'])
#Preserve_regist_date와 Close_date의 항목들 중 1111-11-11 00:00:00 의 값들은 to_datetime으로 변환되지 않음

a=[]
for i in range(len(train['Preserve_regist_date'][train['Preserve_regist_date']=='1111-11-11 00:00:00'].index)):
  a.append(train['Preserve_regist_date'][train['Preserve_regist_date']=='1111-11-11 00:00:00'].index[i])
for i in range(len(train['Close_date'][train['Close_date']=='1111-11-11 00:00:00'].index)):
  a.append(train['Close_date'][train['Close_date']=='1111-11-11 00:00:00'].index[i])  
a.sort()
aa=[]
for i in a:
  if i not in aa:
    aa.append(i)
print(a)
print(len(aa))
aa.reverse()
print(aa)
print(train.shape)
for i in range(len(aa)):
  train=train.drop(aa[i],axis=0)

# 1111-11-11 00:00:00 인 항목들의 위치를 찾아서 그 row들을 모두 제거함

train['Preserve_regist_date']=pd.to_datetime(train['Preserve_regist_date'])
train['Close_date']=pd.to_datetime(train['Close_date'])

# 날짜 형식의 항목들을 모두 datetime으로 바꿔줌

print(train.dtypes)
```

```python
x_train=train.drop('Hammer_price',axis=1)
y_train=train['Hammer_price']
print(y_train.shape)

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(x_train,y_train)
y_pred=clf.predict(test)
```
