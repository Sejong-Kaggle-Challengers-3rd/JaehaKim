```python
print(train.isna().sum())
print(test.isna().sum())

train=train.fillna(0) #nan값을 0으로 채움
test['자격유형']=test['자격유형'].fillna('P')
test['임대보증금']=test['임대보증금'].fillna(0)
test['임대료']=test['임대료'].fillna(0)
test['도보 10분거리 내 지하철역 수(환승노선 수 반영)']=test['도보 10분거리 내 지하철역 수(환승노선 수 반영)'].fillna(0)

print(test.isna().sum()) #제대로 채워졌는지 확인
```

```python
print(train.dtypes)
```

```python
cols=[]
for i in (train.columns):
  if train[i].dtypes=='O':
    cols.append(i)
print(cols)
del cols[6]
del cols[5]
print(cols)
```

```python
for i in(train[train['임대료']=='-'].index):
  train['임대료'][i]=0
for i in(train[train['임대보증금']=='-'].index):
  train['임대보증금'][i]=0

for i in(test[test['임대료']=='-'].index):
  test['임대료'][i]=0
for i in(test[test['임대보증금']=='-'].index):
  test['임대보증금'][i]=0

train[train['임대료']=='-']
train[train['임대보증금']=='-']

train.dtypes
```

```python
for i in range(len(train['임대료'])):
  train['임대료'][i]=int(train['임대료'][i])
  train['임대보증금'][i]=int(train['임대보증금'][i])

for i in range(len(test['임대료'])):
  test['임대료'][i]=int(test['임대료'][i])
  test['임대보증금'][i]=int(test['임대보증금'][i])


print(train.dtypes)

for i in range(len(train['임대료'])):
  if type(train['임대료'][i])!=int:
    print(train['임대료'][i])
```
